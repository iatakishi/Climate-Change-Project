# 2. Data Preparation Section:

2.1 Changing the data types subsection:

```python
#Converting data types of countries_of_the_world by replacing ',' with '.' in numbers and making them float.

countries_of_the_world['Pop. Density (per sq. mi.)'] = countries_of_the_world['Pop. Density (per sq. mi.)'].str.replace(',' , '.').astype(float)
countries_of_the_world['Coastline (coast/area ratio)'] = countries_of_the_world['Coastline (coast/area ratio)'].str.replace(',' , '.').astype(float)
countries_of_the_world['Net migration'] = countries_of_the_world['Net migration'].str.replace(',' , '.').astype(float)
countries_of_the_world['Infant mortality (per 1000 births)'] = countries_of_the_world['Infant mortality (per 1000 births)'].str.replace(',' , '.').astype(float)
countries_of_the_world['Literacy (%)'] = countries_of_the_world['Literacy (%)'].str.replace(',' , '.').astype(float)
countries_of_the_world['Phones (per 1000)'] = countries_of_the_world['Phones (per 1000)'].str.replace(',' , '.').astype(float)
countries_of_the_world['Arable (%)'] = countries_of_the_world['Arable (%)'].str.replace(',' , '.').astype(float)
countries_of_the_world['Crops (%)'] = countries_of_the_world['Crops (%)'].str.replace(',' , '.').astype(float)
countries_of_the_world['Other (%)'] = countries_of_the_world['Other (%)'].str.replace(',' , '.').astype(float)
countries_of_the_world['Climate'] = countries_of_the_world['Climate'].str.replace(',' , '.').astype(float)
countries_of_the_world['Birthrate'] = countries_of_the_world['Birthrate'].str.replace(',' , '.').astype(float)
countries_of_the_world['Deathrate'] = countries_of_the_world['Deathrate'].str.replace(',' , '.').astype(float)
countries_of_the_world['Agriculture'] = countries_of_the_world['Agriculture'].str.replace(',' , '.').astype(float)
countries_of_the_world['Industry'] = countries_of_the_world['Industry'].str.replace(',' , '.').astype(float)
countries_of_the_world['Service'] = countries_of_the_world['Service'].str.replace(',' , '.').astype(float)
```

```python
#Check the data types of countries_of_the_world

countries_of_the_world.dtypes
```

```python
#Converting data types of gender_statistics by replacing '..' with '0' in numbers and making them float.

gender_statistics['2012 [YR2012]'] = gender_statistics['2012 [YR2012]'].str.replace('..', '0', regex=False).astype(float)
gender_statistics['2013 [YR2013]'] = gender_statistics['2013 [YR2013]'].str.replace('..', '0', regex=False).astype(float)
gender_statistics['2014 [YR2014]'] = gender_statistics['2014 [YR2014]'].str.replace('..', '0', regex=False).astype(float)
gender_statistics['2015 [YR2015]'] = gender_statistics['2015 [YR2015]'].str.replace('..', '0', regex=False).astype(float)
gender_statistics['2016 [YR2016]'] = gender_statistics['2016 [YR2016]'].str.replace('..', '0', regex=False).astype(float)
gender_statistics['2017 [YR2017]'] = gender_statistics['2017 [YR2017]'].str.replace('..', '0', regex=False).astype(float)
gender_statistics['2018 [YR2018]'] = gender_statistics['2018 [YR2018]'].str.replace('..', '0', regex=False).astype(float)
gender_statistics['2019 [YR2019]'] = gender_statistics['2019 [YR2019]'].str.replace('..', '0', regex=False).astype(float)
gender_statistics['2020 [YR2020]'] = gender_statistics['2020 [YR2020]'].str.replace('..', '0', regex=False).astype(float)
gender_statistics['2021 [YR2021]'] = gender_statistics['2021 [YR2021]'].str.replace('..', '0', regex=False).astype(float)
```

2.3 Data cleaning and transformation subsection:

```python
#Our tables:
#commodity_trade
#world_happiness_report
#world_happiness_report_2021
#gender_statistics
#countries_of_the_world
#global_emissions
#global_emissions_country
```

```python
# ----------- CLEANING GLOBAL_EMISSIONS -----------------
```

```python
# Pivoting the global_emissions table

global_emissions = global_emissions.melt(id_vars = ['country_name', 'country_code', 'indicator_name', 'indicator_code'],
                                         var_name = 'Year', value_name = 'co2_emissions_kt')
```

```python
# Deleting NaN values of global_emissions

global_emissions = global_emissions.dropna(subset = 'co2_emissions_kt')
```

```python
# Deleting some columns of global_emissions

global_emissions = global_emissions.drop(['indicator_code', 'indicator_name'], axis=1)
```

```python
# ----------- CLEANING GENDER_STATISTICS -----------------
```

```python
# Pivoting the gender_statistics table

gender_statistics = gender_statistics.melt(id_vars = ['series_code', 'country_name', 'country_code', 'series_name'],
                       value_vars = ['2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021'],
                      var_name = 'year', value_name = 'female_male_ratio')
```

```python
# Deleting NaN values of the gender_statistics, female_male_ratio

gender_statistics = gender_statistics.dropna(subset = 'female_male_ratio')
```

```python
# Deleting 0 values of gender_statistics, female_male_ratio

gender_statistics = gender_statistics.drop(gender_statistics[gender_statistics.female_male_ratio == 0].index)
```

```python
# Deleting some columns of gender_statistics table

gender_statistics = gender_statistics.drop(['series_name', 'series_code'], axis=1)
```

```python
# ----------- CLEANING WORLD_HAPPINESS_REPORT_2021 -----------------
```

```python
# Deleting some columns of world_happiness_report_2021

world_happiness_report_2021 = world_happiness_report_2021.drop(['ladder_score_in_dystopia', 'explained_by_log_gdp_per_capita',
                                                                'explained_by_social_support', 
                                                                'explained_by_healthy_life_expectancy', 
                                                                'explained_by_freedom_to_make_life_choices',
                                                                'explained_by_generosity',
                                                                'explained_by_perceptions_of_corruption',
                                                                'dystopia_plus_residual'], axis=1)
```

```python
# ----------- CLEANING COMMODITY_TRADE -----------------
```

```python
# Deleting totals of commodity_trade

commodity_trade = commodity_trade.drop(commodity_trade[commodity_trade.comm_code == 'TOTAL'].index)
```
